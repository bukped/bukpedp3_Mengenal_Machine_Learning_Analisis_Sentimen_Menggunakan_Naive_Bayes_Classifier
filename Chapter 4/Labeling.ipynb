```python
import pandas as pd
import numpy as np

from tqdm import tqdm
from textblob import TextBlob
```

```python
df_translated = pd.read_csv('translated_dataset.csv')
df_raw = pd.read_csv('raw_dataset.csv')
```

```python
df_combine = pd.concat([df_raw, df_translated], axis=1)
```

```python
df_combine.isnull().sum()
```

```python
df_cleaned = df_combine.dropna()

# Tampilkan informasi tentang DataFrame yang sudah dibersihkan
print("Info for Cleaned DataFrame:")
df_cleaned.isnull().sum()
```

```python
df_cleaned['translate'].replace('', np.nan, inplace=True) # replace string kosong dengan Nan
```

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned
```

## Labeling Menggunakan TextBlob

```python
label_texts = []
polarity_text = []
subjectivity_texts = []
scores=[]

for comment in tqdm(df_cleaned['translate']):
    try:
        blob = TextBlob(comment)
        sentimen = blob.sentiment

        # Mendapatkan nilai sentimen
        polarity = sentimen.polarity  # Nilai antara -1 hingga 1 (positif hingga negatif)
        subjectivity = sentimen.subjectivity  # Nilai antara 0 hingga 1 (faktual hingga subjektif)
    except Exception as e:
        print(f"Error at index {index}: {e}")
        print(comment)

    # Menampilkan hasil analisis sentimen
    if polarity > 0:
        sentiment_label = "Positif"
        score=1
    elif polarity < 0:
        sentiment_label = "Negatif"
        score=-1
    else:
        sentiment_label = "Netral"
        score=0
    
    scores.append(score)
    label_texts.append(sentiment_label)
    polarity_text.append(polarity)
    subjectivity_texts.append(subjectivity)
```

```python
df_final = df_cleaned.copy()

df_final['label']=label_texts
df_final['score']=scores
df_final['polarity']=polarity_text
df_final['subjectivity']=subjectivity_texts
    
```
